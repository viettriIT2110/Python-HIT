# Data Types

Để phân tích dữ liệu, điều quan trọng là phải biết loại dữ liệu chúng ta đang xử lý.
Chia các loại dữ liệu thành ba loại chính:
- Numerical (Số)
- Categorical (Dữ liệu phân loại)
- Ordinal (thông thường)
Numerical : dữ liệu số là những con số, và có thể được chia thành hai loại số:
    Dữ liệu rời rạc
        - số bị giới hạn ở số nguyên. Ví dụ: Số lượng ô tô đi qua.
    Dữ liệu liên tục
        - số có giá trị vô hạn. Ví dụ: Giá của một mặt hàng, hoặc kích thước của một mặt hàng
Categorical : Dữ liệu phân loại là các giá trị không thể đo lường được với nhau. Ví dụ: một giá trị màu hoặc bất kỳ giá trị có / không.
Ordinal : Dữ liệu thông thường giống như dữ liệu phân loại, nhưng có thể được đo lường với nhau. Ví dụ: điểm trường trong đó A tốt hơn B và cứ thế.

Khi biết loại dữ liệu của nguồn dữ liệu của bạn, bạn sẽ có thể biết nên sử dụng kỹ thuật nào khi phân tích chúng.

# Mean, Median, and Mode

Trong Machine Learning (và trong toán học) thường có ba giá trị khiến chúng ta quan tâm:
- Mean: Giá Trị Trung bình
- Median : Giá trị trung điểm
- Mode : Giá trị phổ biến nhất
- speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

### Mean
Để tính giá trị trung bình, tìm tổng của tất cả các giá trị và chia tổng cho số lượng giá trị:


(99+86+87+88+111+86+103+87+94+78+77+85+86) / 13 = 89.77
Sử dụng mean()phương pháp NumPy để tìm tốc độ trung bình:

```python
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)

print(x)
```

### Median
Giá trị trung bình là giá trị ở giữa, sau khi bạn đã sắp xếp tất cả các giá trị:


```python
import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

Nếu có hai số ở giữa, chia tổng của các số đó cho hai.


```python
import numpy

speed = [99,86,87,88,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

### MODE
Giá trị mode là giá trị xuất hiện nhiều lần nhất:

Sử dụng mode()phương pháp SciPy để tìm số xuất hiện nhiều nhất:

```python
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = stats.mode(speed)

print(x)
```

# Standard Deviation (Độ lệch chuẩn)
- Độ lệch chuẩn là một số mô tả mức độ lan truyền của các giá trị.

- Độ lệch chuẩn thấp có nghĩa là hầu hết các số gần với giá trị trung bình (trung bình).

- Độ lệch chuẩn cao có nghĩa là các giá trị được trải đều trên một phạm vi rộng hơn.
- speed = [86,87,88,86,87,85,86]
    Độ lệch chuẩn là : 0,9. Có nghĩa là hầu hết các giá trị nằm trong phạm vi 0,9 từ giá trị trung bình, là 86,4.
- speed = [32,111,138,28,59,77,97]
    Độ lệch chuẩn là : 37,85. Có nghĩa là hầu hết các giá trị nằm trong phạm vi 37,85 từ giá trị trung bình, là 77,4.

```python
import numpy

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)

print(x)
```

```python
import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

## Variance (Phương Sai)
Phương sai là một số khác cho biết mức độ lan truyền của các giá trị.

Trong thực tế, nếu bạn lấy căn bậc hai của phương sai, bạn sẽ có độ lệch chuẩn!

Hoặc ngược lại, nếu bạn nhân độ lệch chuẩn với chính nó, bạn sẽ có được phương sai!

Để tính toán phương sai bạn phải làm như sau:

1. Tìm giá trị trung bình:


2. Đối với mỗi giá trị: tìm sự khác biệt từ giá trị trung bình:



3. Đối với mỗi khác biệt: tìm giá trị bình phương:


4. Phương sai là số trung bình của các chênh lệch bình phương này:


5.  công thức để tìm độ lệch chuẩn là căn bậc hai của phương sai:


Độ lệch chuẩn thường được thể hiện bằng biểu tượng Sigma: σ

Phương Sai thường được thể hiện bằng biểu tượng Sigma vuông: σ 2

```python
import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.var(speed)

print(x)
```

# Percentiles

Percentiles được sử dụng trong thống kê để cung cấp cho bạn một số mô tả giá trị mà tỷ lệ phần trăm nhất định của các giá trị thấp hơn.

Ví dụ: Giả sử chúng ta có một loạt các độ tuổi của tất cả những người sống trên đường phố.



75 phần trăm là gì? Câu trả lời là 43, nghĩa là 75% người dân từ 43 tuổi trở xuống.

```python
import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 75)

print(x)
```

Độ tuổi mà 90% người trẻ hơn tuổi là bao nhiêu?

```python
import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 90)

print(x)
```

# Data Distribution
Phân phối dữ liệu

Trước đó trong hướng dẫn này, chúng tôi đã làm việc với một lượng rất nhỏ dữ liệu trong các ví dụ của mình, chỉ để hiểu các khái niệm khác nhau.

Trong thế giới thực, các tập dữ liệu lớn hơn nhiều, nhưng có thể khó thu thập dữ liệu trong thế giới thực, ít nhất là ở giai đoạn đầu của một dự án.

Làm thế nào chúng ta có thể có được bộ dữ liệu lớn?

Để tạo các tập dữ liệu lớn để thử nghiệm, chúng tôi sử dụng mô-đun NumPy của Python, đi kèm với một số phương thức để tạo các tập dữ liệu ngẫu nhiên, ở mọi kích thước.


- Tạo một mảng chứa 250 số float ngẫu nhiên trong khoảng từ 0 đến 5:

```python
import numpy 

x = numpy.random.uniform(0.0, 5.0, 250)

print(x)
```

## Histogram
Để trực quan hóa tập dữ liệu, chúng ta có thể vẽ biểu đồ với dữ liệu chúng ta đã thu thập.

Chúng tôi sẽ sử dụng mô-đun Python Matplotlib để vẽ biểu đồ:

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 250)

plt.hist(x, 5)
plt.show()
```

Một mảng chứa 250 giá trị không được coi là rất lớn, nhưng bây giờ bạn đã biết cách tạo một bộ giá trị ngẫu nhiên và bằng cách thay đổi các tham số, bạn có thể tạo tập dữ liệu lớn như bạn muốn.

Tạo một mảng với 100000 số ngẫu nhiên và hiển thị chúng bằng biểu đồ có 100 thanh:

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100)
plt.show()
```
