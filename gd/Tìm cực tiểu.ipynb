## x**2 + 5 * sin(x)

```python
import math
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# hàm tính f(x)
def f(x):
    return x**2 + 5*np.sin(x)
```

```python
# hàm tính đạo hàm tại x
def grad(x):
    return 2*x + 5* np.cos(x)
```

```python
# lặp khi nào đạo hàm có độ lớn đủ nhỏ
def myGD1(eta, x0):
    x = [x0]
    for it in range(100):
        x_new = x[-1] - eta*grad(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return (x, it)
```

```python
(x1, it1) = myGD1(.1, -5)
(x2, it2) = myGD1(.1, 5)

y1 = [f(x) for x in x1]
plt.plot(x1,y1)
plt.show()

print('Solution x1 = %f, cost = %f, obtained after %d iterations'%(x1[-1], f(x1[-1]), it1))

y2 = [f(x) for x in x2]
plt.plot(x2,y2)
plt.show()

print('Solution x2 = %f, cost = %f, obtained after %d iterations'%(x2[-1], f(x2[-1]), it2))
```
